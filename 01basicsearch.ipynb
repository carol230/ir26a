# Ejercicio 01: Búsqueda Básica

**Objetivo:** En este ejercicio vamos a implementar una búsqueda básica dentro de un corpus pequeño de textos.



```python
import os

# Usamos la misma carpeta '100libros' que ya se descargó con download_data.ipynb
ruta = os.path.join(os.getcwd(), '100libros')

archivos = os.listdir(ruta)
print(archivos)

contenidos = {}

for archivo in archivos:
    with open(os.path.join(ruta, archivo), 'r', encoding='utf-8') as f:
        contenido = f.read()
        contenidos[archivo] = contenido
        print(f"Contenido del archivo {archivo}:\n{contenido[:5]}\n")
```

```python
import time

query = "Mary"

start_time = time.time()
resultados = []
for file_name in contenidos.keys():
    if query in contenidos[file_name]:
        resultados.append(file_name)

print(f"Archivos que contienen la palabra '{query}': {resultados}")
end_time = time.time()
print(f"Tiempo de búsqueda: {end_time - start_time:.4f} segundos")
```

```python
def buscar(query: str, contenidos: dict) -> list:
    resultados = []
    for file_name in contenidos.keys():
        if query in contenidos[file_name]:
            resultados.append(file_name)
    return resultados
```

```python
# Se usa 'Don Quijote.txt' como libro base para extraer el vocabulario
# (reemplaza a pg84.txt que era del corpus original de Gutenberg)
with open(os.path.join(ruta, 'Don Quijote.txt'), 'r', encoding='utf-8') as f:
    contenido = f.read()
    palabras = set(contenido.split())

len(palabras)
```

```python
import time

# MEJORA: se usa una muestra de 500 palabras en lugar del vocabulario completo (~40,000).
# Razón: el corpus ahora son 100 libros (antes eran pocos archivos pequeños),
# buscar todas las palabras tardaría varios minutos. La muestra es suficiente
# para demostrar que la búsqueda lineal es lenta comparada con el índice invertido.
muestra = list(palabras)[:500]

start_time = time.time()
for palabra in muestra:
    buscar(palabra, contenidos)
end_time = time.time()

tiempo_muestra = end_time - start_time
tiempo_estimado_total = tiempo_muestra * (len(palabras) / len(muestra))

print(f"Tiempo para {len(muestra)} palabras: {tiempo_muestra:.4f} segundos")
print(f"Vocabulario total: {len(palabras)} palabras")
print(f"Tiempo estimado para búsqueda completa: {tiempo_estimado_total:.2f} segundos")
```
