## Índice Invertido

Un indice invertido es una estructura de datos en la que se detalla en qué documentos aparece cada una de las palabras del vocabulario dentro del corpus. 

```python
import os

# MEJORA: ruta relativa construida con os.path.join.
# Razón: evita rutas hardcodeadas de otro equipo; os.getcwd() devuelve
# la carpeta donde está corriendo el notebook, por lo que '100libros'
# debe existir dentro de esa carpeta.
path = os.path.join(os.getcwd(), '100libros')

files = os.listdir(path)
len(files)
```

```python
contenidos = {}
indice_invertido = {}

for archivo in files:
    # MEJORA: os.path.join en lugar de path + archivo
    with open(os.path.join(path, archivo), 'r', encoding='utf-8') as f:
        contenido = f.read()
        palabras = contenido.split()
        vocabulario = set(palabras)
        for palabra in vocabulario:
            if palabra in indice_invertido:
                indice_invertido[palabra].append(archivo)
            else:
                indice_invertido[palabra] = [archivo]
        print('Archivo:', archivo + ' - Número de palabras únicas:', len(vocabulario))
        print('Numero de palabras del inidice invertido: ', len(indice_invertido))
```

```python
len(indice_invertido)
```

```python
import pandas as pd

df = pd.DataFrame(list(indice_invertido.items()), columns=['Palabra', 'Archivos'])
df
```

```python
def buscar(query):
    return indice_invertido.get(query, [])
```

```python
# Ejemplos de búsqueda en el índice invertido
# MEJORA: en lugar de imprimir las 841,418 entradas (lo que causaba el
# KeyboardInterrupt anterior), se muestra cómo buscar palabras específicas
# y se imprime solo una muestra del índice para verificar que funciona.

print(buscar('gobiernan'))
print(buscar('amor'))
print(buscar('muerte'))

# Muestra las primeras 5 entradas del índice como verificación
print("\n--- Primeras 5 entradas del índice ---")
for palabra in list(indice_invertido.keys())[:5]:
    print(f"'{palabra}': {buscar(palabra)}")
```
